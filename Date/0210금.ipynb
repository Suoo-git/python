```python
import requests
res = requests.get('https://newsapi.org/v2/top-headlines?country=kr&apiKey=9f5baf7d9f3f42879a20d7d19d9886e4')

result = res.json() #

result
result.get('articles')

art = result.get('articles')
for arts in art:
    published = arts.get('publishedAt')
    year = published[:4]
    month = published[5:7]
    day = published[8:10]
    print(f"{year}년 {month}월 {day}일")

    # find함수 이용
    idx1 = published.find('-')
    y = published[:idx1]
    idx2 = published.find('-', idx1)
    m = published[idx1 + 1 : idx2]
    d = published[idx2 +1 : published.find('T')]
    print(y, m, d)
```
